```python
# we webscrape and compute significant scores on sentences to determine their importance
# important senteces are then picked out to 'summarise' the text in question
import urllib3
from bs4 import BeautifulSoup
```

```python
url = 'https://www.bbc.co.uk/news/world-europe-66440335'
http = urllib3.PoolManager()
resp = http.request("GET", url)
```

```python
page = resp.data.decode('utf8', 'ignore') # this is raw html
soup = BeautifulSoup(page, "html.parser")
```

```python
# get first instance of <title>
print(soup.title)
#extract text
print(soup.title.string)
```

```python
print(soup.prettify())
```

```python
# find all paragraph tags
p_tags = soup.find_all("p")
p_tags[:15]
```

```python
soup.find_all("p", attrs={"class":"ssrcss-1q0x1qg-Paragraph e1jhz7w10"})
```

```python
# convert to strings
p_strings = [p.string for p in p_tags]
p_strings[:10]
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
sentences = word_tokenize(''.join(p_strings))
sentences[:10]
```
